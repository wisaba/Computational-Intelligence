```python
import random as rd
```

```python
class TSP:
    def __init__(self):
        self.sample = [i + 1 for i in range(51)]
        self.t = []
        self.population = []
        self.costs = [[]]
        for i in range(20):
            rd.shuffle(self.sample)
            t = self.sample.copy()
            self.population.append(t)
        with open("TSP51.txt", "r") as file:
            for line in file:
                values = line.split()
                self.costs.append([int(values[1]), int(values[2])])
        


    def distance(self, x1, x2):
        return ((x1[0] - x2[0]) ** 2 + (x1[1] - x2[1]) ** 2) ** 0.5

    def fitness(self, x):
        cost = 0
        for i in range(len(x) - 1):
            cost += self.distance(self.costs[x[i]], self.costs[x[i + 1]])
        cost += self.distance(self.costs[x[0]], self.costs[x[len(x) - 1]])
        return cost

    def cross_over(self, x1, x2):
        t1 = rd.randint(0, 49)  
        t2 = rd.randint(t1 + 1, 50)

        child = x1[t1:t2] 

        i = 0
        for city in x2:
            if city not in child:  
                while i < t1 or i >= t2:
                    if i == 50:
                        i = 0
                    else:
                        i += 1
                child.insert(i, city)

        return child

    def mutation(self, x):
        t1 = rd.randint(0, 49)
        t2 = rd.randint(0, 49)
        x[t1], x[t2] = x[t2], x[t1]  
        return x

    def Both_C_and_M(self, x):
        population = []
        k = 0
        for i in range(len(x)):
            for j in range(len(x)):
                k += 1
                temp = self.cross_over(x[i], x[j])
                if k % 10 == 0:
                    population.append(self.mutation(temp))
                else:
                    population.append(temp)
        return population[:15]
```

```python
def main():
    t = TSP()
    population = t.population
    for generation in range(1000):
        population = [[t.fitness(i), i] for i in population]
        population = sorted(population)
        elitisms = population[:5]
        population2 = [elitisms[j][1] for j in range(len(elitisms))]
        after_mut_cross = [i[1] for i in population[5:11]]
        after_mut_cross = t.Both_C_and_M(after_mut_cross)
        population2 += after_mut_cross
        population = population2.copy()
        print(f"The Best Selectiob in the {generation} generation:", population[0], "And Cost Is:", t.fitness(population[0]))
    print("The Best the End:", population[0], "And  Cost Is:", t.fitness(population[0]))
    
```

```python
if __name__ == "__main__":
    main()
```
